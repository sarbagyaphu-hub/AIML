```python
# Task 1: Unit Converter

def convert_units(conversion_type, value):
    """
    Convert a value between supported units.

    Parameters:
    conversion_type (str): Type of conversion to perform.
        Supported options:
        - 'm_to_ft'   : meters to feet
        - 'ft_to_m'   : feet to meters
        - 'kg_to_lbs' : kilograms to pounds
        - 'lbs_to_kg' : pounds to kilograms
        - 'l_to_gal'  : liters to gallons
        - 'gal_to_l'  : gallons to liters
    value (float): The numeric value to convert.

    Returns:
    float: Converted value.
    """
    if conversion_type == "m_to_ft":
        return value * 3.28084
    elif conversion_type == "ft_to_m":
        return value / 3.28084
    elif conversion_type == "kg_to_lbs":
        return value * 2.20462
    elif conversion_type == "lbs_to_kg":
        return value / 2.20462
    elif conversion_type == "l_to_gal":
        return value * 0.264172
    elif conversion_type == "gal_to_l":
        return value / 0.264172
    else:
        raise ValueError("Unsupported conversion type.")


print("Unit Converter")
print("Choose a conversion type:")
print("1. Meters to Feet")
print("2. Feet to Meters")
print("3. Kilograms to Pounds")
print("4. Pounds to Kilograms")
print("5. Liters to Gallons")
print("6. Gallons to Liters")

try:
    choice = input("Enter your choice (1-6): ").strip()
    value = float(input("Enter the value to convert: "))

    if choice == "1":
        result = convert_units("m_to_ft", value)
        print(f"{value} meters = {result:.4f} feet")
    elif choice == "2":
        result = convert_units("ft_to_m", value)
        print(f"{value} feet = {result:.4f} meters")
    elif choice == "3":
        result = convert_units("kg_to_lbs", value)
        print(f"{value} kilograms = {result:.4f} pounds")
    elif choice == "4":
        result = convert_units("lbs_to_kg", value)
        print(f"{value} pounds = {result:.4f} kilograms")
    elif choice == "5":
        result = convert_units("l_to_gal", value)
        print(f"{value} liters = {result:.4f} gallons")
    elif choice == "6":
        result = convert_units("gal_to_l", value)
        print(f"{value} gallons = {result:.4f} liters")
    else:
        print("Invalid choice. Please enter a number from 1 to 6.")

except ValueError:
    print("Invalid input. Please enter a numeric value.")
except Exception as e:
    print("An error occurred:", e)
```

```python
# Task 2: Mathematical Operations on a List of Numbers

def find_sum(numbers):
    """
    Returns the sum of all numbers in the list.

    Parameters:
    numbers (list): A list of numeric values.

    Returns:
    float: Sum of the numbers.
    """
    return sum(numbers)


def find_average(numbers):
    """
    Returns the average of all numbers in the list.

    Parameters:
    numbers (list): A list of numeric values.

    Returns:
    float: Average of the numbers.
    """
    return sum(numbers) / len(numbers)


def find_maximum(numbers):
    """
    Returns the maximum number in the list.

    Parameters:
    numbers (list): A list of numeric values.

    Returns:
    float: Maximum value in the list.
    """
    return max(numbers)


def find_minimum(numbers):
    """
    Returns the minimum number in the list.

    Parameters:
    numbers (list): A list of numeric values.

    Returns:
    float: Minimum value in the list.
    """
    return min(numbers)


print("Mathematical Operations on a List")
print("Choose an operation:")
print("1. Sum")
print("2. Average")
print("3. Maximum")
print("4. Minimum")

try:
    choice = input("Enter your choice (1-4): ").strip()
    user_input = input("Enter numbers separated by spaces: ").strip()

    if not user_input:
        raise ValueError("The list cannot be empty.")

    numbers = [float(num) for num in user_input.split()]

    if choice == "1":
        result = find_sum(numbers)
        print("Sum =", result)
    elif choice == "2":
        result = find_average(numbers)
        print("Average =", result)
    elif choice == "3":
        result = find_maximum(numbers)
        print("Maximum =", result)
    elif choice == "4":
        result = find_minimum(numbers)
        print("Minimum =", result)
    else:
        print("Invalid choice. Please enter a number from 1 to 4.")

except ValueError as e:
    print("Invalid input:", e)
except Exception as e:
    print("An error occurred:", e)
```

```python
# Exercise 4.2: List Manipulation (All in One)

# 1. Extract Every Other Element
def extract_every_other(lst):
    """Returns every other element starting from the first."""
    return lst[::2]


# 2. Slice a Sublist
def get_sublist(lst, start, end):
    """Returns sublist from start to end (inclusive)."""
    return lst[start:end+1]


# 3. Reverse a List Using Slicing
def reverse_list(lst):
    """Returns reversed list."""
    return lst[::-1]


# 4. Remove First and Last Elements
def remove_first_last(lst):
    """Removes first and last element."""
    return lst[1:-1]


# 5. Get First n Elements
def get_first_n(lst, n):
    """Returns first n elements."""
    return lst[:n]


# 6. Get Last n Elements
def get_last_n(lst, n):
    """Returns last n elements."""
    return lst[-n:]


# 7. Reverse Skip (every second from second-last)
def reverse_skip(lst):
    """Returns every second element from second-last backwards."""
    return lst[-2::-2]


# ------------------ TESTING ------------------

lst = [1, 2, 3, 4, 5, 6]

print("Original List:", lst)

print("1. Every other:", extract_every_other(lst))
print("2. Sublist (2 to 4):", get_sublist(lst, 2, 4))
print("3. Reversed:", reverse_list(lst))
print("4. Remove first & last:", remove_first_last(lst))
print("5. First 3 elements:", get_first_n(lst, 3))
print("6. Last 2 elements:", get_last_n(lst, 2))
print("7. Reverse skip:", reverse_skip(lst))
```

```python
# Exercise 4.3: Nested List (All in One)

# 1. Flatten a Nested List
def flatten(lst):
    """Flattens a 2D nested list into a single list."""
    result = []
    for item in lst:
        if isinstance(item, list):
            result.extend(item)
        else:
            result.append(item)
    return result


# 2. Access Nested Element
def access_nested_element(lst, indices):
    """Access element using list of indices."""
    element = lst
    for i in indices:
        element = element[i]
    return element


# 3. Sum of All Elements (Deep)
def sum_nested(lst):
    """Returns sum of all elements in nested list."""
    total = 0
    for item in lst:
        if isinstance(item, list):
            total += sum_nested(item)
        else:
            total += item
    return total


# 4. Remove Specific Element
def remove_element(lst, elem):
    """Removes all occurrences of elem from nested list."""
    result = []
    for item in lst:
        if isinstance(item, list):
            result.append(remove_element(item, elem))
        else:
            if item != elem:
                result.append(item)
    return result


# 5. Find Maximum Element
def find_max(lst):
    """Finds max value in nested list."""
    maximum = float('-inf')
    for item in lst:
        if isinstance(item, list):
            maximum = max(maximum, find_max(item))
        else:
            maximum = max(maximum, item)
    return maximum


# 6. Count Occurrences
def count_occurrences(lst, elem):
    """Counts occurrences of elem in nested list."""
    count = 0
    for item in lst:
        if isinstance(item, list):
            count += count_occurrences(item, elem)
        else:
            if item == elem:
                count += 1
    return count


# 7. Deep Flatten (any depth)
def deep_flatten(lst):
    """Flattens list of any depth."""
    result = []
    for item in lst:
        if isinstance(item, list):
            result.extend(deep_flatten(item))
        else:
            result.append(item)
    return result


# 8. Average of Nested List
def average_nested(lst):
    """Returns average of all elements."""
    flat = deep_flatten(lst)
    return sum(flat) / len(flat)


# ------------------ TESTING ------------------

nested = [[1, 2], [3, [4, 5]], 6]

print("Original Nested List:", nested)

print("1. Flatten:", flatten([[1, 2], [3, 4], [5]]))
print("2. Access element [1,1,0]:", access_nested_element(nested, [1,1,0]))
print("3. Sum:", sum_nested(nested))
print("4. Remove 2:", remove_element([[1, 2], [3, 2], [4, 5]], 2))
print("5. Max:", find_max(nested))
print("6. Count 2:", count_occurrences([[1,2],[2,3],[2,4]], 2))
print("7. Deep Flatten:", deep_flatten(nested))
print("8. Average:", average_nested([[1, 2], [3, 4], [5, 6]]))
```
