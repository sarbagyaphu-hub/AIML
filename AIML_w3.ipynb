```python
# ===============================
# TASK 1: MCP NEURON IMPLEMENTATION
# ===============================

# -------- AND FUNCTION --------
def MCP_Neurons_AND(X1, X2, T=2):
    assert len(X1) == len(X2)

    state_neuron = []

    for i in range(len(X1)):
        total = X1[i] + X2[i]

        if total >= T:
            state_neuron.append(1)
        else:
            state_neuron.append(0)

    return state_neuron


# -------- OR FUNCTION --------
def MCP_Neurons_OR(X1, X2, T=1):
    assert len(X1) == len(X2)

    state_neuron = []

    for i in range(len(X1)):
        total = X1[i] + X2[i]

        if total >= T:
            state_neuron.append(1)
        else:
            state_neuron.append(0)

    return state_neuron


# -------- INPUT DATA --------
X1 = [0, 0, 1, 1]
X2 = [0, 1, 0, 1]

# -------- RUN AND --------
and_result = MCP_Neurons_AND(X1, X2)
print("AND Output:", and_result)

# -------- RUN OR --------
or_result = MCP_Neurons_OR(X1, X2)
print("OR Output:", or_result)


# ===============================
# EXTRA: XOR USING IF-ELSE LOGIC
# ===============================
def MCP_XOR_logic(X1, X2):
    result = []

    for i in range(len(X1)):
        if X1[i] != X2[i]:
            result.append(1)
        else:
            result.append(0)

    return result

xor_result = MCP_XOR_logic(X1, X2)
print("XOR Output:", xor_result)
```

```python
# ==========================================
# TASK 2: PERCEPTRON FOR 0 vs 1 CLASSIFICATION
# ==========================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# -------------------------------
# 1. LOAD THE DATASET
# -------------------------------
# Change the file path if needed
df_0_1 = pd.read_csv("mnist_0_and_1.csv")

# Extract features and labels
X = df_0_1.drop(columns=["label"]).values   # 784 pixel features
y = df_0_1["label"].values                  # labels: 0 or 1

print("Feature matrix shape:", X.shape)
print("Label vector shape:", y.shape)

# -------------------------------
# 2. VISUALIZE THE DATASET
# -------------------------------
images_0 = X[y == 0]
images_1 = X[y == 1]

fig, axes = plt.subplots(2, 5, figsize=(10, 5))

if len(images_0) < 5 or len(images_1) < 5:
    print("Error: Not enough images in images_0 or images_1 to plot 5 images.")
else:
    for i in range(5):
        # plot digit 0
        axes[0, i].imshow(images_0[i].reshape(28, 28), cmap="gray")
        axes[0, i].set_title("Label: 0")
        axes[0, i].axis("off")

        # plot digit 1
        axes[1, i].imshow(images_1[i].reshape(28, 28), cmap="gray")
        axes[1, i].set_title("Label: 1")
        axes[1, i].axis("off")

    plt.suptitle("First 5 Images of 0 and 1 from MNIST Subset")
    plt.tight_layout()
    plt.show()

# -------------------------------
# 3. INITIALIZE WEIGHTS AND BIAS
# -------------------------------
weights = np.zeros(X.shape[1])   # one weight for each pixel
bias = 0
learning_rate = 0.1
epochs = 100

print("Initial weights shape:", weights.shape)
print("Initial bias:", bias)

# -------------------------------
# 4. DECISION / ACTIVATION FUNCTION
# -------------------------------
def decision_function(X, weights, bias):
    """
    Compute predicted labels using step activation.
    """
    predictions = np.dot(X, weights) + bias
    y_pred_all = np.where(predictions >= 0, 1, 0)
    return y_pred_all

# -------------------------------
# 5. TRAIN THE PERCEPTRON
# -------------------------------
def train_perceptron(X, y, weights, bias, learning_rate=0.1, epochs=100):
    """
    Train perceptron using perceptron learning rule.
    """
    for epoch in range(epochs):
        correct_predictions = 0

        for i in range(len(X)):
            output = np.dot(X[i], weights) + bias
            y_pred = 1 if output >= 0 else 0

            error = y[i] - y_pred

            if error != 0:
                weights = weights + learning_rate * error * X[i]
                bias = bias + learning_rate * error
            else:
                correct_predictions += 1

        accuracy = correct_predictions / len(X)
        print(f"Epoch {epoch + 1}/{epochs} - Accuracy: {accuracy:.4f}")

        # optional early stopping
        if accuracy == 1.0:
            print("Training converged early.")
            break

    return weights, bias, accuracy

# -------------------------------
# 6. TRAIN THE MODEL
# -------------------------------
weights, bias, accuracy = train_perceptron(
    X, y, weights, bias, learning_rate=learning_rate, epochs=epochs
)

print("Final Accuracy:", accuracy)

# -------------------------------
# 7. EVALUATE THE MODEL
# -------------------------------
y_pred = decision_function(X, weights, bias)
final_accuracy = np.mean(y_pred == y)

print(f"Final Accuracy after evaluation: {final_accuracy:.4f}")

# -------------------------------
# 8. VISUALIZE MISCLASSIFIED IMAGES
# -------------------------------
misclassified_idx = np.where(y_pred != y)[0]

if len(misclassified_idx) > 0:
    fig, axes = plt.subplots(2, 5, figsize=(10, 5))
    for ax, idx in zip(axes.flat, misclassified_idx[:10]):
        ax.imshow(X[idx].reshape(28, 28), cmap="gray")
        ax.set_title(f"Pred: {y_pred[idx]}, True: {y[idx]}")
        ax.axis("off")

    plt.suptitle("Misclassified Images")
    plt.tight_layout()
    plt.show()
else:
    print("All images were correctly classified!")
```
